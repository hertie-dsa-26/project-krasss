```python
import requests
import pandas as pd
```

```python
# Define parameters
api_key = "71166a96655234262760b0975365279ba3b63e04"
years = range(2013, 2024)
vars = "NAME,DP05_0001E,DP05_0003PE,DP05_0002PE,DP05_0017E,DP02_0065PE,DP02_0066PE,DP02_0061PE,DP05_0077PE,DP05_0078PE,DP05_0080PE,DP05_0071PE,DP03_0062E"
```

```python
all_data = []

for year in years:
    # URL for ACS 5-Year Data Profile
    url = f"https://api.census.gov/data/{year}/acs/acs5/profile?get={vars}&for=county:*&key={api_key}"
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        df = pd.DataFrame(data[1:], columns=data[0])
        df['year'] = year # Add year column for tracking
        all_data.append(df)

# Combine and save
final_df = pd.concat(all_data, ignore_index=True)
final_df.to_csv("combined_census_final.csv", index=False)
```

```python
# update column names
final_df = final_df.rename(columns={
    "DP05_0001E":  "total_population",
    "DP05_0002PE": "pct_male",
    "DP05_0003PE": "pct_female",
    "DP05_0017E":  "median_age",
    "DP02_0065PE": "pct_bachelors_plus",
    "DP02_0066PE": "pct_graduate_degree",
    "DP02_0061PE": "pct_less_than_hs",
    "DP05_0077PE": "pct_white",
    "DP05_0078PE": "pct_black",
    "DP05_0080PE": "pct_asian",
    "DP05_0071PE": "pct_hispanic",
    "DP03_0062E":  "median_household_income"
})

print(final_df.columns.tolist())
final_df.head()
```
